```python
# 3.2.1 Visualize Sigmoid Function
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(-10, 10, 0.1)
y = 1 / (1+np.exp(-x))

# 그래프 그리기
plt.plot(x, y)
plt.show()
```

```python
# 3.2.3
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
    return np.array(x > 0, dtype=np.int)

x = np.arange(-5, 5, 0.1) # typing error
y = step_function(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.1) # y축 범위
plt.show()
```

```python
# 3.2.4
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1, 1, 2])
sigmoid(x)
```

```python
x = np.arange(-5, 5, 0.1)
y = sigmoid(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# 3.2.5
x = np.arange(-5, 5, 0.1)
y1 = sigmoid(x)
y2 = step_function(x)

plt.plot(x,y1)
plt.plot(x,y2, linestyle="--")
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# 3.2.7 ReLU
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5, 5, 0.1)
y = relu(x)
y1 = sigmoid(x)
y2 = step_function(x)

plt.plot(x,y, color="red")
plt.plot(x,y1, linestyle="-.")
plt.plot(x,y2, linestyle="--")
plt.ylim(-0.1, 3)
plt.show()
```
