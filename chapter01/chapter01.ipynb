```python
# 1.4.2
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
        
    def hello(self):
        print("Hello "+ self.name +"!")
        
    def goodbye(self):
        print("Good-bye "+ self.name +"!")

        
m = Man("David")
m.hello()
m.goodbye()
```

```python
# 1.6.1
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.show()
```

```python
# 1.6.2
# 데이터 준비
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # 점선으로 그리기
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
